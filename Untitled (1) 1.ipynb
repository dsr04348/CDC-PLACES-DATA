```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file

data = pd.read_csv(file_path)

# Select the columns for the health conditions
health_conditions = ['ACCESS2_CrudePrev', 'ARTHRITIS_CrudePrev', 'BINGE_CrudePrev']

# Plot bar charts for each health condition
plt.figure(figsize=(18, 6))
for i, condition in enumerate(health_conditions, start=1):
    # Sort the data by the prevalence of the current health condition
    sorted_data = data.sort_values(by=condition, ascending=False).head(10)
    
    plt.subplot(1, 3, i)
    plt.bar(sorted_data['CountyName'], sorted_data[condition], color='skyblue')
    plt.title(f'Top 10 Counties - {condition}')
    plt.xlabel('County')
    plt.ylabel('Prevalence')
    plt.xticks(rotation=90)
    plt.tight_layout()

plt.show()
```

```python
import pandas as pd

# Load the dataset
df = data

# Select the top 10 counties based on total population
top_10_counties = df.groupby('CountyName')['TotalPopulation'].max().nlargest(10).index

# Filter the dataset for the top 10 counties
df_top_10 = df[df['CountyName'].isin(top_10_counties)]

# Define the health conditions to visualize
health_conditions = ['ACCESS2_CrudePrev', 'ARTHRITIS_CrudePrev', 'BINGE_CrudePrev', 'BPHIGH_CrudePrev']

# Create bar charts for each health condition
for condition in health_conditions:
    # Calculate mean prevalence for each county
    mean_prevalence = df_top_10.groupby('CountyName')[condition].mean()
    
    # Plot bar chart
    mean_prevalence.plot(kind='bar', figsize=(10, 6), title=f'Prevalence of {condition}', ylabel='Prevalence (%)')
    plt.xlabel('County')
    plt.xticks(rotation=45)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, Dropdown

# Assuming `data` is your DataFrame
df = data

# Function to plot scatter plot between values of health_condition_1 for county_name_1 and values of health_condition_2 for county_name_2
def plot_scatter_between_health_conditions(county_name_1, county_name_2, health_condition_1, health_condition_2):
    county_data_1 = df[df['CountyName'] == county_name_1]
    county_data_2 = df[df['CountyName'] == county_name_2]

    # Plot
    plt.figure(figsize=(12, 8))
    plt.scatter(county_data_1[health_condition_1], county_data_2[health_condition_2])
    
    plt.xlabel(f'{county_name_1}: {health_condition_1}')
    plt.ylabel(f'{county_name_2}: {health_condition_2}')
    plt.title(f'Scatter Plot between {county_name_1}: {health_condition_1} and {county_name_2}: {health_condition_2}')
    plt.grid(True)
    plt.show()

# Get unique county names
county_names = df['CountyName'].unique()

# Get columns ending with 'CrudePrev'
crude_prev_columns = [col for col in df.columns if col.endswith('CrudePrev')]

# Dropdown widget for county selection
county_dropdown_1 = Dropdown(options=county_names, description='County 1:')
county_dropdown_2 = Dropdown(options=county_names, description='County 2:')

# Dropdown widget for selecting health conditions
condition_dropdown_1 = Dropdown(options=crude_prev_columns, description='Health Condition 1:')
condition_dropdown_2 = Dropdown(options=crude_prev_columns, description='Health Condition 2:')

# Interact function to update the plot based on dropdown selections
interact(plot_scatter_between_health_conditions, 
         county_name_1=county_dropdown_1, 
         county_name_2=county_dropdown_2,
         health_condition_1=condition_dropdown_1, 
         health_condition_2=condition_dropdown_2)



```

```python
import geopandas as gpd
import folium
from folium.plugins import MarkerCluster
from ipywidgets import interact, Dropdown

# Load the shapefile containing the geometries of Indiana census tracts
indiana_tracts = gpd.read_file(r'C:\Users\deepa\Downloads\Census_Tract_Boundaries_2020\census_tracts_of_indiana_uscb_2020_tl_2020_18_tract2.shp')

# Assuming `data` is your DataFrame
df = data

# Get columns ending with 'CrudePrev'
crude_prev_columns = [col for col in df.columns if col.endswith('CrudePrev')]

# Dropdown widget for selecting health condition
condition_dropdown_1 = Dropdown(options=crude_prev_columns, description='Health Condition 1:')

# Function to plot the map
def plot_health_condition_map(health_condition_1):
    # Merge census tract data with health condition data
    merged_data = indiana_tracts.merge(df, left_on='geoid20', right_on='TractFIPS', how='left')
    
    # Create a Folium map
    m = folium.Map(location=[39.7685, -86.1581], zoom_start=7)
    
    # Add a marker cluster
    marker_cluster = MarkerCluster().add_to(m)
    
    # Add markers for each census tract
    for idx, row in merged_data.iterrows():
        folium.Marker(
            location=[row.geometry.centroid.y, row.geometry.centroid.x],
            popup=f'{row["CountyName"]}: {row[health_condition_1]}',
            icon=None,
        ).add_to(marker_cluster)
    
    # Display the map
    display(m)

# Interact function to update the map based on dropdown selection
interact(plot_health_condition_map, health_condition_1=condition_dropdown_1)
```
